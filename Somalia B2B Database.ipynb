```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
file_path = "Somalia firmographic Sample Kaggle_Count.xlsx"
df = pd.read_excel(file_path)

# Display basic info and missing values
df.info()
missing_values = df.isnull().sum().sort_values(ascending=False)
print("Missing Values per Column:\n", missing_values)

# Visualize missing data
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title("Missing Values Heatmap")
plt.show()

# Sales Volume Distribution
if 'SalesVolumeDollars' in df.columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(df['SalesVolumeDollars'].dropna(), kde=True, bins=10, color='blue')
    plt.title("Distribution of Sales Volume (Dollars)")
    plt.xlabel("Sales Volume ($)")
    plt.ylabel("Frequency")
    plt.grid(True)
    plt.show()

# Count businesses by city
if 'City' in df.columns:
    city_counts = df['City'].value_counts().dropna()
    plt.figure(figsize=(8, 5))
    city_counts.plot(kind='bar', color='teal')
    plt.title("Number of Businesses by City")
    plt.xlabel("City")
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.grid(axis='y')
    plt.show()

# Display relevant business information
print(df[['CompanyName', 'City', 'SalesVolumeDollars', 'EmployeesHere', 'PhoneOrMobile']].head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset (assuming the file is already available)
file_path = "Somalia firmographic Sample Kaggle_Count.xlsx"
df = pd.read_excel(file_path)

# Fill missing values with placeholders (as you did previously)
df.fillna({
    'Email': 'Not Provided',
    'Website': 'Not Available',
    'Address1': 'Unknown',
    'Province': 'Unknown',
    'Region': 'Unknown'
}, inplace=True)

# Sort companies by SalesVolumeDollars (assuming 'SalesVolumeDollars' represents revenue)
df_sorted = df[['CompanyName', 'SalesVolumeDollars']].dropna().sort_values(by='SalesVolumeDollars', ascending=False)

# Add a rank column based on revenue
df_sorted['Rank'] = df_sorted['SalesVolumeDollars'].rank(ascending=False)

# Create a line chart of company ranks vs. sales volume
plt.figure(figsize=(10, 6))
plt.plot(df_sorted['Rank'], df_sorted['SalesVolumeDollars'], marker='o', linestyle='-', color='b')
plt.title("Company Ranks by Revenue")
plt.xlabel("Rank")
plt.ylabel("Revenue (Sales Volume in Dollars)")
plt.grid(True)
plt.xticks(ticks=range(len(df_sorted)), labels=df_sorted['CompanyName'], rotation=90)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset (assuming the file is already available)
file_path = "Somalia firmographic Sample Kaggle_Count.xlsx"
df = pd.read_excel(file_path)

# Fill missing values with placeholders (as you did previously)
df.fillna({
    'Email': 'Not Provided',
    'Website': 'Not Available',
    'Address1': 'Unknown',
    'Province': 'Unknown',
    'Region': 'Unknown'
}, inplace=True)

# Sort companies by SalesVolumeDollars (assuming 'SalesVolumeDollars' represents revenue)
df_sorted = df[['CompanyName', 'SalesVolumeDollars']].dropna().sort_values(by='SalesVolumeDollars', ascending=False)

# Add a rank column based on revenue
df_sorted['Rank'] = df_sorted['SalesVolumeDollars'].rank(ascending=False)

# Create a horizontal bar chart for better visualization of ranks and revenues
plt.figure(figsize=(10, 8))
sns.barplot(x='SalesVolumeDollars', y='CompanyName', data=df_sorted, palette='viridis')
plt.title("Company Ranks by Revenue")
plt.xlabel("Revenue (Sales Volume in Dollars)")
plt.ylabel("Company Name")
plt.grid(True, axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
